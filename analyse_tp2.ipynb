```python
import numpy as np 
np.set_printoptions(threshold=10000,suppress=True) 
import pandas as pd 
import warnings 
import matplotlib.pyplot as plt 
```

```python
warnings.filterwarnings('ignore') 
data_villes = pd.read_csv('./villes.csv', sep=';') 
X_villes = data_villes.iloc[:, 1:13].values 
labels_villes = data_villes.iloc[:, 1].values 


# print("Aperçu des données :")
# print(data_villes)
```

```python
# on garde les axes qui expliquent 80% à 90% de la variance totale
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler_villes = StandardScaler()
X_scaled_villes = scaler_villes.fit_transform(X_villes)
pca_villes = PCA(n_components=2)
X_pca_villes = pca_villes.fit_transform(X_scaled_villes)

# calcul de la variance expliquée cumulée
# cumulative_variance = np.cumsum(pca_villes.explained_variance_ratio_)

# affichage des résultats
# for i, variance in enumerate(cumulative_variance):
#    print(f"Axes 0 à {i+1} : {variance*100:.2f}% de variance expliquée")
    
# cela affiche : 
"""
Axes 0 à 1 : 87.27% de variance expliquée
Axes 0 à 2 : 98.99% de variance expliquée
Axes 0 à 3 : 99.46% de variance expliquée
Axes 0 à 4 : 99.72% de variance expliquée
Axes 0 à 5 : 99.84% de variance expliquée
Axes 0 à 6 : 99.89% de variance expliquée
Axes 0 à 7 : 99.93% de variance expliquée
Axes 0 à 8 : 99.96% de variance expliquée
Axes 0 à 9 : 99.98% de variance expliquée
Axes 0 à 10 : 99.99% de variance expliquée
Axes 0 à 11 : 100.00% de variance expliquée
Axes 0 à 12 : 100.00% de variance expliquée
"""

# on garde donc les deux premiers axes car ils expliquent 
# a plus de 80% la variance, ce qui est suffisant

import matplotlib 
plt.scatter(X_pca_villes[:, 0], X_pca_villes[:, 1]) 
for label, x, y in zip(labels_villes, X_pca_villes[:, 0], X_pca_villes[:, 1]): 
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
plt.title("PCA - villes.csv")
plt.show()  
```
